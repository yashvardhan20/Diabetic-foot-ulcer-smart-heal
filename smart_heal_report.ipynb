<h4><div style="text-align: center;">Summary</div></h4>


Leg ulcers are deep asymetrically shaped sores that tend to be slow in healing. A wound below the knee that has not responded to treatment within 3 months or has not healed fully after 12 months is considered a chronic leg ulcer (ref). Ulcers can have a significant impact on the quality of life of patients. Pain, cosmetic considerations and mobility contributing to both loss of productivity and in many cases psychological distress further complicating treatment and recovery. (ref 10) Patients presenting with leg ulcers or the risk factors are on the rise in Australia. Data from the BEACH program was used to show that from 2000 to 2014 there was a significant increase participating GP's treating chronic skin ulcers at a rate of 7 per 1000 encounters. The data shows that the increase is due to older Australians seeing GP's. Of the patients presenting, leg ulcers were most prevalent at 43%, foot ulcers were 10% and venous 5%, a further 15% were unspecified. The prevalence of leg ulcers was the same in both male and female patient encounters. Smart Heal aims to use modern handheld devices and advances in AI to process wound data providing Doctors with easy access to treatments, dignostic tooling and patient progress. We have been tasked with providing an algorthimic basis for treatment subsequent to the use of the mobile app to scan the wound and on entry of patient history and other key diagnostics such as smoking status. We propose that a dignostic matrix be created from readily available high quality research publications. This matrix would be compiled with factors such as size of wound, periwound condition and shape, age of patient, risk factors and others. This will be coupled to established treatments and used as a dataset for a standard ML procedure. The treatment options will result from a prediction from the precedure on application of new patient details consistent with the matrix.

<h4><div style="text-align: center;">Introduction</div></h4>

The most common cause for leg ulcers is chronic venous insufficiency at 70% (ref) followed by arterial disease up 10% (ref) others include diabetes or vasculitis or combination. 

#### Patient history considerations in determining patient treatment

1. Onset and duration - any preceding trauma or recurrence
2. Progress of Ulcer - Slow would point to Arterial Ulcer, Venous would likely progress faster
3. Assesment of Pain - Arterial Disease, Vasculitis and hypertension produce more painful ulcers. Venous ulcers don't always show symptoms in absence of secondary infection. Neuropathic ulcers may or may not present with tingling or numbness.
4. Associated features - Pain on standing associated with hyperpigmentation or dermatitis of lower legs favours a venous desease whereas pain on walking is associated with Arterial disease.
5. Medical History - Diabetes, hypertension, pervious surgery, recurring bacterial infection, deep vein thrombosis, connective tissue disorders, rheumatoid arthritis, varicose veins (autoimmune disorders) are risk factors for Ulcers.
6. Drug History - Many topical medications can affect wound healing.
7. Family History - Blood disorders
8. Addictions - Smoking, alcohol

#### Other Features In Treatment Determination
1. Location size and shape and how these change on follow up (progress)
2. Patient mobility, limited mobility or abnormalities can affect effectiveness of treatment
3. Results of swab and or biopsy of ulcer in order to determine correct course of antibiotics particularly important for patients with Diabetes
4. Lab checks for undiagnosed systemic diseases (autoimmune conditions) 

#### Treatment
Wound Debridgement eliminating superficial or necrotic tissue. 
Debridgement is not recommended in cases of severe arterial insufficiency.
Treat any infection. 
Dress the wound tailoring the dressing based on the location, size, shape, depth and presence of slough
Frequency of dressing change is important, dressing should never be saturated and never dry and adherent to the wound
There are a number of dressings absorbent, plastic films, hydrocolloids and others
Hydrogel sheets and Hydrocolloid dressings are useful for dry wounds with minimal exudate
Low adherent absorbent dressings are appropriate for shallow wounds with minimal to low exudate and minimal trauma at each change
Alginate dressings are suitable for wounds with high exudate and particularly useful in stemming bleeding
Semipermeable adhesive films useful in shallow to flat wounds with minimal exudate.
Chlorhexidine impregnated dressings help to reduce bacterial colonisation but some patients can be allergic to Chlorhexidine
Compression therapies
Pressure off loading particularly in diabetic ulcers
